This notebook loads the raw dataset, inspects its structure, and documents baseline conditions. No cleaning or modification happens here.

## 1. Import Libraries
Importing pandas, numpy, matplotlib, and seaborn for data loading and initial inspection. Nothing else is needed here.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the CSV
Loading from data/raw/. Confirming the shape matches expectations.

```python
df = pd.read_csv('../data/raw/hospital_readmissions_30k.csv')
print("Dataset Shape:", df.shape)
```

## 3. Print Column Names and Data Types
Checking if any numerical column loaded as an object (string). That would break calculations later.

```python
print("--- Column Names ---")
print(df.columns.tolist())

print("\n--- Data Types ---")
print(df.dtypes)
```

## 4. Head and Tail
Checking the first 5 rows and last 5 rows to ensure the data looks sensible end to end.

```python
display(df.head())
display(df.tail())
```

## 5. Statistical Summary
Running describe()

```python
display(df.describe(include='all'))
```

## 6. Check for Missing Values
Checking for null values. This documents our baseline cleanliness.

```python
print("--- Missing Values Count ---")
print(df.isnull().sum())
```

## 7. Check for Duplicate Rows
Checking for exact duplicate rows in the dataset. Expected: 0.

```python
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

## 8. Value Counts on Categorical Columns
Running value counts on specific categorical columns where dirty/undocumented values might live.

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_cols:
    print(f"\n--- Value Counts for {col} ---")
    print(df[col].value_counts(dropna=False))
```
