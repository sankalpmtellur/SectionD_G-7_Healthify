# 05 Final Load Preparation

**Objective:** Prepare KPI tables and aggregated datasets for dashboarding in Tableau. This notebook turns cleaned analytical data into dashboard fuel.

```python
import pandas as pd
import os

df = pd.read_csv('../data/processed/cleaned_data.csv')
os.makedirs('../tableau', exist_ok=True)
df.head()
```

## KPI 1: Overall Readmission Rate

```python
readmission_rate = pd.DataFrame({
    'metric': ['overall_readmission_rate'],
    'value': [df['is_readmitted'].mean() * 100]
})
readmission_rate
```

## KPI 2: Readmission by Age Group

```python
age_kpi = df.groupby('age_group').agg(
    total_patients=('patient_id','count'),
    readmission_rate=('is_readmitted','mean')
).reset_index()
age_kpi['readmission_rate'] = age_kpi['readmission_rate'] * 100
age_kpi
```

## KPI 3: Readmission by Risk Tier

```python
risk_kpi = df.groupby('risk_tier').agg(
    total_patients=('patient_id','count'),
    readmission_rate=('is_readmitted','mean')
).reset_index()
risk_kpi['readmission_rate'] = risk_kpi['readmission_rate'] * 100
risk_kpi
```

## KPI 4: Readmission by Stay Category

```python
stay_kpi = df.groupby('stay_category').agg(
    total_patients=('patient_id','count'),
    readmission_rate=('is_readmitted','mean')
).reset_index()
stay_kpi['readmission_rate'] = stay_kpi['readmission_rate'] * 100
stay_kpi
```

## KPI 5: Chronic Condition Impact

```python
condition_kpi = df.groupby(['diabetes','hypertension']).agg(
    total_patients=('patient_id','count'),
    readmission_rate=('is_readmitted','mean')
).reset_index()
condition_kpi['readmission_rate'] = condition_kpi['readmission_rate'] * 100
condition_kpi
```

## KPI 6: Discharge Destination Analysis

```python
discharge_kpi = df.groupby('discharge_destination').agg(
    total_patients=('patient_id','count'),
    readmission_rate=('is_readmitted','mean')
).reset_index()
discharge_kpi['readmission_rate'] = discharge_kpi['readmission_rate'] * 100
discharge_kpi
```

## Export KPI Files for Tableau

```python
readmission_rate.to_csv('../tableau/readmission_rate.csv', index=False)
age_kpi.to_csv('../tableau/age_kpi.csv', index=False)
risk_kpi.to_csv('../tableau/risk_kpi.csv', index=False)
stay_kpi.to_csv('../tableau/stay_kpi.csv', index=False)
condition_kpi.to_csv('../tableau/condition_kpi.csv', index=False)
discharge_kpi.to_csv('../tableau/discharge_kpi.csv', index=False)

print('All KPI files exported successfully.')
```
