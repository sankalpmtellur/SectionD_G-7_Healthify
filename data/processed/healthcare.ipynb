```python
Male = health.loc[health['Gender'] == 'Male'].shape[0]
Female = health.loc[health['Gender'] == 'Female'].shape[0]
if pd.isna(Male):
    Male = 0
if pd.isna(Female):
    Female = 0
if Male == 0 and Female == 0:
    print("No data available for plotting.")
else:
    plt.pie([Male, Female], labels=['Male', 'Female'], autopct='%.2f%%')
    plt.title('Gender Distribution')
    plt.show()
```
